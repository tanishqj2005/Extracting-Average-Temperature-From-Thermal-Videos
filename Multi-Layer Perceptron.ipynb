# This file involves code for training a Multi-Layer Perceptron Model on the MNIST Data which would be used while extracting the number from the videos

-> Importing the Necessary Libraries:

```python
import pandas as pd
import numpy as np

from tensorflow.keras.utils import to_categorical
import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

-> Reading the MNIST Data

```python
df = pd.read_csv("/content/drive/MyDrive/Trisem 3/PR and ML/Mini Project/mnist_train.csv")
```

-> Shuffling the Data

```python
df = df.sample(frac = 1)
```

-> Creating Feature and Label Arrays

```python
X = df[[i for i in df.columns[1:]]].values
Y = df[df.columns[0]].values.ravel()

Y = to_categorical(Y)
```

-> Creating the Training and Testing Datasets:

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size = 0.3)
```

-> Creating the Multi-Layer Perceptron Model:

-> We experimented on various architectures and concluded that this one was giving best results:

```python
model_mlp = Sequential()
model_mlp.add(Dense(512, input_dim=784, activation='relu'))
model_mlp.add(Dense(512, input_dim=784, activation='relu'))
model_mlp.add(Dense(256, activation='relu'))
model_mlp.add(Dense(128, activation='relu'))
model_mlp.add(Dense(10, activation='softmax'))

model_mlp.compile('adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'],)
```

-> Training the Model:

```python
history = model_mlp.fit(X, Y, epochs=15, batch_size=64,validation_split = 0.3)
```

-> Making Predictions on the Test Data:

```python
Y_pred = [np.argmax(i) for i in model_mlp.predict(X_test)]
```

-> Evaluating the Predictions:

```python
from sklearn.metrics import accuracy_score

accuracy_score(Y_test.argmax(axis=1),Y_pred)
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(Y_test.argmax(axis=1),Y_pred)
```

```python
from sklearn.metrics import classification_report

print(classification_report(Y_test.argmax(axis=1),Y_pred))
```

-> Saving the Model Weights:

```python
model_mlp.save_weights("/content/drive/MyDrive/Trisem 3/PR and ML/Mini Project/model-mlp.ckpt")
```
